# OOOs class wrapping the algo

```python
import numpy as np

class LogisticRegression:

    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        

    def fit(self, X, y):
        n_samples, n_features = X.shape

# initializing zero to a array of 0s for shape - example (3,) or (n_features,)
# This means that the weights or coeff of each feature is initialised to 0 
# Bias is a single value not an array which is also initalised to 0

 
        
        self.weights = np.zeros(n_features)
        self.bias = 0


# Gradeint descent

# Iterate over a loop 1000 times
# each time update weights and bias by derivation mentioned above in a figure

# Example:
# data:(200, 3)
# Labels: (200, 1)
# Weights:(3, 1)
  
        for _ in range(self.n_iters):
# dot multiple (3,1) with (200, 3) will result in (200,1) - predictions - or log of odd or y --> one for each observation
# y  = mx + c
# x - weights
# c = bias

# apply sigmoid on y or log of odds to get the predicted probablities
# The aim of each iteration is to reduce the difference between y_actual and y_predicted 
    
                
           
            linear_model = np.dot(X, self.weights) + self.bias
            
            y_predicted = self._sigmoid(linear_model)
    
# Transpose data from (200, 3) to (3, 200) so we can multiply with weight (3,1) - by the dot multiply rule
# This will result in (3,1) weights/ coeffs - one for each feature -- representing the aggregate
# slope of the cost function across all observations

# Bias is updated by the formula above

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

# Update weights and bias by the formula applying learning rate
    
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
        return self.weights, self.bias  
    
    '''
    predict takes the final weight and bias and fit into the linear equation - y
    apply sigmoid on y / linear model is performed to get predictions
    return in array format
    '''
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self._sigmoid(linear_model)
        return np.array(y_predicted)
    '''
    a threshold is defined to based on the requirement of senitivity or recall or precision and predictions are 
    either classed as 1 or 0 
    '''    
    
    def predict_class(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self._sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return np.array(y_predicted_cls)
    '''
    helper function to return sigmoid
    
    '''
    
    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))   
    
```

```python
'''
To calculate log loss from scratch, we need to include the MinMax function (see below). 
Numpy implements this for us with np.clip(). This will clip each value of the y_pred array so to avoid the nan cost  
coming from any value very close to 0 or 1 - log(0) or log(1-1) which will result in nan value

'''


def logloss(y_test, y_pred, eps=1e-15):
    p = np.clip(y_pred, eps, 1 - eps)
    class2_cost = (1-p)*np.log(1-p)
    class1_cost = y_test*np.log(p)
    cost = class2_cost + class1_cost
    return -1* cost.sum()/len(y_test)    
```

# Testing the algo with a dataset

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
bc = datasets.load_breast_cancer()
X = bc['data'] # bc.data
X.shape
```

```python
y = bc['target'] #bc.target
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 7)
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
```

weights  = np.zeros(30)

```python
regressor = LogisticRegression(lr=0.001, n_iters=1000)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
y_pred
```

```python
weights, bias = regressor.fit(X_train, y_train)
print(f'weight/coeff: {weights}')
print(f'bias/intecept: {bias}')
```

```python
logloss(y_test, y_pred, eps=1e-15)
```

```python
y_pred_class = regressor.predict_class(X_test)
y_pred_class
```

```python
def accuracy(y_test, y_pred_class):
    accuracy =  np.sum(y_test == y_pred_class)/ len(y_test)
    return accuracy
```

```python
print('accuracy score:',accuracy(y_test, y_pred_class))
```
