# Multiple Linear Regression Algo

```python
import numpy as np
X = np.array([[0.038,0.051,0.062,0.022,-0.044],[-0.002, -0.045, -0.051, -0.026, -0.008],[0.085, 0.051, 0.044, -0.006, -0.046],[-0.089, -0.045, -0.012, -0.037, 0.012],[0.005, -0.045, -0.036, 0.022, 0.004],[-0.093, -0.045, -0.041, -0.019, -0.069],[-0.045, 0.051, -0.047, -0.016, -0.04],[0.064, 0.051, -0.002, 0.067, 0.091]])
print(X)
```

```python
y = np.array([151,75,141,206,135,97,138,63])
y
```

```python
n_samples, n_features = X.shape
```

```python
weights = np.zeros(n_features)
bias = 0
lr = 1.6
lamb = 0.5
```

for i in range(5):
        print(f'weights:{weights}')
        print(f'bias:{bias}')
        y_pred = np.dot(X, weights) + bias 
        dw = (-1 / n_samples) * np.dot(X.T, (y_pred - y))
        db = (-1 / n_samples) * np.sum(y_pred - y)
        weights += lr * dw
        bias += lr * db    

```python
for i in range(5):
        print(f'weights:{weights}')
        print(f'bias:{bias}')
        y_pred = np.dot(X, weights) + bias 
        dw = (-1 / n_samples) * (np.dot(X.T, (y_pred - y)) + 2*lamb*weights)
        db = (-1 / n_samples) * np.sum(y_pred - y)
        weights += lr * dw
        bias += lr * db   
```

# With Lasso L1

```python
lamb = 1
for i in range(1000):
        print(f'weights:{weights}')
        print(f'bias:{bias}')
        y_pred = np.dot(X, weights) + bias 
        dw = (-1 / n_samples) * (np.dot(X.T, (y_pred - y)) + lamb*np.sum(np.abs(weights)))
        db = (-1 / n_samples) * np.sum(y_pred - y)
        cost = (1/ n_samples)*sum([val**2 for val in (y-y_pred)]) + 2lamb*np.sum(np.abs(weights))
        print(f'error:{cost}')
        weights += lr * dw
        bias += lr * db
```

# With Ridge L2 and cost function

```python
lamb = 1
for i in range(1000):
        print(f'weights:{weights}')
        print(f'bias:{bias}')
        y_pred = np.dot(X, weights) + bias 
        dw = (-1 / n_samples) * (np.dot(X.T, (y_pred - y)) + 2*lamb*weights)
        db = (-1 / n_samples) * np.sum(y_pred - y)
        cost = (1/ n_samples)*sum([val**2 for val in (y-y_pred)]) + 2*lamb*np.sum(weights)
        print(f'error:{cost}')
        weights += lr * dw
        bias += lr * db 
```
